```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("DailyDelhiClimateTrain.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
print(df.describe())
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10, 5))
plt.plot(df['date'], df['meantemp'],marker='o', linestyle='-', color='b',  label="Mean Temp (C)")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.title("Daily Mean Temperature in Delhi")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['humidity'], kde=True, color='g')
plt.xlabel("Humidity (%)")
plt.ylabel("Frequency")
plt.title("Distribution of Humidity Levels")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['meantemp'], kde=True ,color='red')
plt.xlabel("Temperature (C)")
plt.ylabel("Frequency")
plt.title("Temperature Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['wind_speed'], kde=True, color='green')
plt.xlabel("Wind Speed")
plt.ylabel("Frequency")
plt.title("Wind Speed Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['meanpressure'], kde=True, color='purple')
plt.xlabel("Mean Pressure")
plt.ylabel("Frequency")
plt.title("Mean Pressure Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(df['meantemp'], df['humidity'], color='red')
plt.xlabel("Temperature (C)")
plt.ylabel("Humidity (%)")
plt.title("Temperature vs Humidity")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(df['meantemp'], df['wind_speed'], color='green')
plt.xlabel("Temperature (C)")
plt.ylabel("Wind Speed")
plt.title("Temperature vs Wind Speed")
plt.show()
```

```python
plt.scatter(df['humidity'], df['wind_speed'], color='purple')
plt.xlabel("Humidity (%)")
plt.ylabel("Wind Speed")
plt.title("Humidity vs Wind Speed")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(y=df['meantemp'], color="orange")
plt.title("Temperature Distribution")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(y=df['humidity'], color="blue")
plt.title("Humidity Distribution")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(y=df['wind_speed'], color="green")
plt.title("Wind Speed Distribution")
plt.show()
```

```python
sns.pairplot(df, diag_kind="kde")
plt.show()
```

```python
plt.plot(df['date'], df['meantemp'], color='red')
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.title("Daily Mean Temperature in Delhi")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['meantemp'], color='red', label='Temperature (C)')
plt.plot(df['date'], df['humidity'], color='blue', label='Humidity (%)')
plt.plot(df['date'], df['wind_speed'], color='green', label='Wind Speed')

plt.xlabel("Date")
plt.ylabel("Values")
plt.title("Temperature, Humidity & Wind Speed Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
df['rolling_temp'] = df['meantemp'].rolling(window=7).mean()

plt.figure(figsize=(20, 10))
plt.plot(df['date'], df['meantemp'], color='gray', alpha=0.5, label="Actual Temp")
plt.plot(df['date'], df['rolling_temp'], color='red', label="7-day Rolling Mean")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.title("Temperature Trend with Rolling Mean")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

df['wind_category'] = pd.cut(df['wind_speed'], bins=[0, 5, 10, 20], labels=['Low', 'Medium', 'High'])
wind_counts = df['wind_category'].value_counts()

plt.pie(wind_counts, labels=wind_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'orange', 'red'])
plt.title("Wind Speed Categories")
plt.show()
```
